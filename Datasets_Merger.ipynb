```python
# ============================================================
# NSE MULTI-YEAR CSV MERGER (FINAL VERSION)
# ============================================================

import pandas as pd
import os

def merge_nse_folder(folder_path, output_filename="NIFTY_Combined.csv"):
    """
    Merges all NSE CSV files inside a folder into one clean dataset.
    Removes duplicate dates and sorts chronologically.
    """

    all_files = [os.path.join(folder_path, f)
                 for f in os.listdir(folder_path)
                 if f.endswith(".csv")]

    if not all_files:
        print("No CSV files found in folder.")
        return None

    print("Found files:")
    for f in all_files:
        print(" -", os.path.basename(f))

    df_list = []

    for file in all_files:
        df = pd.read_csv(file, encoding="utf-8-sig")

        # Clean column names
        df.columns = df.columns.str.strip()

        # Convert Date column
        df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%Y')

        df_list.append(df)

    # Merge all files
    merged_df = pd.concat(df_list, ignore_index=True)

    # Remove duplicate dates (important for overlapping years)
    merged_df = merged_df.drop_duplicates(subset='Date')

    # Sort chronologically (oldest → newest)
    merged_df = merged_df.sort_values('Date')

    # Reset index
    merged_df.reset_index(drop=True, inplace=True)

    # Save merged file
    merged_df.to_csv(output_filename, index=False)

    print("\nMerged successfully!")
    print("Output file:", output_filename)
    print("Total rows:", len(merged_df))
    print("Date range:", merged_df['Date'].min(), "to", merged_df['Date'].max())

    return merged_df


# ============================================================
# HOW TO USE
# ============================================================

# Put all yearly NSE CSV files inside this folder
folder_path = "NSE_data"   # <-- change this to your folder name

merged_df = merge_nse_folder(folder_path, "NIFTY_n_Years.csv")
```
